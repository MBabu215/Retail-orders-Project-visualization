```python
#Univariate analysis- is a statistical method used to analyze the characteristics of a single variable at a time
#Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the CSV
df = pd.read_csv('retail_orders/orders.csv')
limit = df.head(10)
df.head(30)
limit
```

```python
df.columns
```

```python
#2. Print Summary Statistics (Console Output)
print("--- Raw Summary Statistics (df.describe()) ---")
print(df.describe())
```

```python
# Choose a variable for univariate analysis â€” let's use 'cost price' as an example
costp = 'cost price'
# 1. Histogram
sns.histplot(df['cost price'], kde=True, color='skyblue')
plt.title(f'Histogram of {'cost price'}')
plt.xlabel('cost price')
plt.ylabel('count')
plt.show()
```

```python

# 2. Box Plot
sns.boxplot(x=df['cost price'], color='lightgreen')
plt.title(f'Box Plot of {'cost price'}')
plt.xlabel('cost price')
plt.show()
```

```python
#2. Bivariate Analysis â€” Relationship Between Two Variables
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='cost price', y='List Price', data=df)
plt.title('Cost Price vs. List Price')
plt.xlabel('Cost Price')
plt.ylabel('List Price')
plt.show()
```

```python
#B. Box Plot â€” Categorical vs Numeric
sns.boxplot(x='Quantity', y='Discount Percent', data=df, palette='Set2')
plt.title('Discount Percent vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Discount Percent')
plt.show()
```

```python
#Grouped Bar Plot â€” Categorical vs Categorical

import seaborn as sns
import matplotlib.pyplot as plt

data = df

# Your plot
sns.countplot(x='Category', hue='Sub Category', data=data, palette='Set1')

# Title and axis labels
plt.title("Category by Sub Category")
plt.xlabel("Category")
plt.ylabel("Count")

# Move legend to the right side
plt.legend(title='Sub Category', bbox_to_anchor=(1.05, 1), loc='upper left')

# Make sure layout doesn't cut off the legend
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#3. Multivariate Analysis â€” 3+ Variables
 # Multivariate Scatter Plot with Hue (color by category)
sns.scatterplot(data=df,x='cost price',y='List Price',hue='Category',# change to 'Region' or 'Segment' if you prefer
    size='Quantity',alpha=0.7
)
plt.title("Cost vs List Price by Category (size = Quantity)")
plt.show()
```

```python
# Multivariate Pair Plot
#Pair Plot â€” Matrix of Scatter Plots-essential when we are exploring our data set and trying to become familiar with it.
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant numeric columns for pairplot
num_cols = ['cost price', 'List Price', 'Quantity', 'Discount Percent']

# Generate the pairplot
sns.pairplot(df[num_cols])
plt.suptitle("Pair Plot of Retail Order Metrics", y=1.02)
plt.show()
```

```python
# Multivariate Pair Plot
#Pair Plot with hue â€” Matrix of Scatter Plots-essential when we are exploring our data set and trying to become familiar with it.
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns for analysis
num_cols = ['cost price', 'List Price', 'Quantity', 'Discount Percent']

# Generate pairplot with hue
sns.pairplot(df[num_cols + ['Category']], hue='Category')
plt.suptitle("Pair Plot of Retail Metrics by Category", y=1.02)
plt.show()
```

```python
#Heatmap of Correlation Matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for correlation analysis
numerical_cols = ['cost price', 'List Price', 'Quantity', 'Discount Percent']
corr_matrix = df[numerical_cols].corr()#calculates pairwise correlation coefficients.

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title("Heatmap of Correlation Matrix for Retail Metrics")
plt.show()
#annot=True-displays the numeric values inside the heatmap.
#cmap='coolwarm'-gives a clear color range from negative to positive correlation.
#linewidths=0.5 helps separate the cells
#fmt=".2f" limits the number formatting to two decimal places.
```

```python
#Visualizing Categorical Data
import matplotlib.pyplot as plt

# Set up the figure
plt.figure(figsize=(12, 6))

# Pie Chart 1 - Category
plt.subplot(1, 2, 1)
df['Category'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=140,
    title='Sales Distribution by Category',
    ylabel=''  # Hide y-label
)
plt.axis('equal')
```
