```python
#Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the built-in 'tips' dataset
data = sns.load_dataset("tips")
data.head()
```

```python
#Univariate Analysis — One Variable at a Time
#A. Histogram — Distribution of Numeric Data

sns.histplot(data['total_bill'], kde=True, color='skyblue')
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Count")
plt.show()
```

```python
#B. Box Plot — Shows Outliers and Spread
sns.boxplot(y=data['tip'], color='lightgreen')
plt.title("Boxplot of Tips")
plt.ylabel("Tip ($)")
plt.show()

```

```python
sns.boxplot(y=data['total_bill'], color='lightblue')
plt.title("Boxplot of Total bill")
plt.ylabel("total bill ($)")
plt.show()
```

```python
#C. Bar Plot (Count Plot) — Categorical Frequency

sns.countplot(x='day', data=data, palette='pastel')
plt.title("Number of Orders per Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
#2. Bivariate Analysis — Relationship Between Two Variables
#A. Scatter Plot — Numeric vs Numeric
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.title("Tip vs Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
#B. Box Plot — Categorical vs Numeric
sns.boxplot(x='day', y='total_bill', data=data, palette='Set2')
plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
#C. Grouped Bar Plot — Categorical vs Categorical
sns.countplot(x='day', hue='sex', data=data, palette='Set1')
plt.title("Orders per Day by Gender")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
#3. Multivariate Analysis — 3+ Variables
#A. Hue and Style in Scatter Plot
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', data=data)
plt.title("Tip vs Bill by Gender and Smoking")
plt.show()
```

```python
#B. Pair Plot — Matrix of Scatter Plots
sns.pairplot(data[['total_bill', 'tip', 'size', 'sex']], hue='sex')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
#C. Heatmap — Correlation Matrix
correlation = data[['total_bill', 'tip', 'size']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Visualizing Categorical Data
#A. Pie Chart — Category Proportions
data['day'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Orders by Day")
plt.ylabel("")
plt.show()
```

```python
#B. Violin Plot — Distribution + Density
sns.violinplot(x='day', y='total_bill', data=data, palette='muted')
plt.title("Total Bill Distribution per Day")
plt.show()
```

```python
#C. Swarm Plot — Exact Data Points
sns.swarmplot(x='day', y='tip', data=data, palette='Accent')
plt.title("Tips by Day (Exact Values)")
plt.show()
```
