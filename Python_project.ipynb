```python
from sqlalchemy import create_engine, Column, Integer, String, TIMESTAMP, DECIMAL, NUMERIC
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
import pandas as pd
df.values.tolist()

# Create engine
database_url = "postgresql+psycopg2://postgres:admin@localhost:5432/postgres"
engine = create_engine(database_url)

# Create Base class
Base = declarative_base()

# Define ORM table
class RetailOrder(Base):
    __tablename__ = "Retail_Dataa"

    order_id = Column(Integer, primary_key=True, autoincrement=True)
    order_date = Column(TIMESTAMP, nullable=False)
    customer_id = Column(Integer, nullable=False)
    category = Column(String(50), nullable=False)
    sub_category = Column(String(50), nullable=True)
    product_id = Column(Integer, nullable=False)
    selling_price = Column(NUMERIC(10, 2), nullable=False)
    discount_percent = Column(DECIMAL(5, 2), nullable=True)
    discount = Column(NUMERIC(10, 2), nullable=True)
    ship_mode = Column(String(50), nullable=True)
    city = Column(String(100), nullable=True)
    state = Column(String(100), nullable=True)
    region = Column(String(100), nullable=True)
    postal_code = Column(String(100), nullable=True)
    country = Column(String(100), nullable=True)
    segment = Column(String(100), nullable=True)
    profit = Column(NUMERIC(10, 2), nullable=True)

# Create table in database
Base.metadata.create_all(engine)

# Example DataFrame (replace this with your actual data)
data = {
    "order_id": [10],
    "order_date": ["2024-01-01 10:00:00"],
    "customer_id": [1],
    "category": ["Office Supplies"],
    "sub_category": ["Paper"],
    "product_id": [1001],
    "selling_price": [20.50],
    "discount_percent": [5.00],
    "discount": [1.03],
    "ship_mode": ["Second Class"],
    "city": ["New York"],
    "state": ["NY"],
    "region": ["East"],
    "postal_code": ["10001"],
    "country": ["USA"],
    "segment": ["Consumer"],
    "profit": [5.75],
}

# Convert to DataFrame
df = pd.DataFrame(data)
# Append data to table
df.to_sql("Retail_Dataa", con=engine, if_exists="append", index=False)
```
